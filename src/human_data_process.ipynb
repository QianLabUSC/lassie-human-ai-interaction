```python
import pickle

file_path = "tune/human_data/2020_hh_trials_all.pickle"
with (open(file_path, "rb")) as openfile:
    data = pickle.load(openfile)

```

```python
# print(data.layout_name.unique())
# we use counter_circuit for human data finetune
import json
import ast
from llm.utils import write_to_file

new_df = data.loc[data['layout_name'] == "counter_circuit"]

out_file_path = "tune/human_data/2020_hh_trials_counter_circuit.json"
# print(content)

# read new_df to json
json.dump(new_df.to_dict(orient='records'), open(out_file_path, "w"), indent=4)
```

```python

jdata = json.loads(open(out_file_path).read())
final_log = []

for frame in jdata:
    log = {}
    # we only focus on agent 1 action where action are not stay
    agent_1_action = json.loads(frame["joint_action"])[0]
    if agent_1_action != [0, 0]:
        
        log["action"] = agent_1_action
        state = json.loads(frame["state"])
        log["agent"] = state["players"][0]
        log["other_agent"] = state["players"][1]
        layout = json.loads(frame["layout"])
        log["world_state"] = state["objects"]
        log["grid"] = json.loads(frame["layout"])
        final_log.append(log)
        print(log)

        
```

```python
# then append subtask to each log
from data import rawdata_add_subtask
final_log = rawdata_add_subtask(final_log)
json.dump(final_log, open(out_file_path, "w"), indent=4)
```
